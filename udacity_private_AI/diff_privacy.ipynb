```python
#the imprts
import torch
import numpy as np

```

```python
# make 10 hospitals and 10000 data points with 10 labels each of them have
num_hospitals = 10 
num_lables = 10
num_data = 10000
```

```python
#making the predictions data
predictions = (np.random.rand(num_hospitals,num_data)*num_lables).astype(int).transpose(1,0)
predictions[:,0]
```

```python
#making the noisy array 
new_labels = list()
#for each of the label we are adding the noise 
for an_image in predictions:
    label_counts = np.bincount(an_image,minlength=num_lables)  #created the label counts 

    epsilon = 0.1 #addition of the noise 
    beta = 1/epsilon

    for i in range(len(label_counts)):
        label_counts[i] += np.random.laplace(0, beta, 1)  #added the noise for the label counts

    new_label = np.argmax(label_counts)

    new_labels.append(new_label) 
```
