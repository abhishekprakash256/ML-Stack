```python
#imports 
import torch 
import numpy as np
```

```python
#creating the values 
num_labels = 10 # the label for one of the data 
num_teachers = 10 #as 10 hospitals 
num_exapmles = 10000 #size of the dataset 
```

## Generate the Fake data 

```python
predictions = (np.random.rand(num_teachers, num_exapmles) * num_labels).astype(int).transpose(1,0)
print(predictions.shape)
```

##  Add the noise to the data set

```python
new_labels = list()
for an_image in predictions:
    label_counts = np.bincount(an_image, minlength=num_labels)

    epsilon = 0.1
    beta = 1/epsilon
    
    for i in range(len(label_counts)):
        label_counts[i] = np.random.laplace(0,beta,1)
    new_label = np.argmax(label_counts)
    new_labels.append(new_label)

```
